# PyCity Schools Analysis

## summary and analysis

This database contains information from 15 schools with a total of 39,170 students. The database also has more specific information about each school and student, for example, grade, gender, math score, reading score, school budget, size, and type. 

In the data frame "district summary"  we analyze the data from the schools in general, in which we can see that 74.98% of all students pass math and 85.80% pass reading, but only 65.17% of the students pass math and reading. We can also see the average score in each subject, total budget, total students, and total schools.In the second data frame "School summary" you can find the same information as the 1 data frame but in specific for all schools in addition to the budget per student and school type. 

The following data frames show the top and bottom 5 schools per overall passing percentage.  In the top 5 schools, more than 90% of their students pass the two subjects and the difference between the average math and reading score is by decimals (in all cases math scores are lower). In difference with the lowest performing schools only less than 54% of the students pass both subjects, the average math scores are significantly lower than the average reading scores, less than 67% of the students pass math. One thing that we can conclude from this comparison is that to for the majority of students, math may be more difficult than reading and that surprisingly the bottom schools have more budget per student than the top schools, but these schools are smaller in size and charter-type which means that maybe they can put more attention to each student, instead of the bottom schools that are district-type and with a bigger amount of students.

The two successive data frames show the average score for each subject by grade in each school. the average didn´t have a significate change during the grades, the difference was less than 2.

In the spending ranges analysis per student, you can see that the less they spend, the higher the scores get and the same thing happens in the scores by school size analysis in which it´s shown that the smallest the school is the better the students do in the scores. In the end, the analysis of the scores by school type was done, and the charter schools had significantly higher scores and percentages of passing than the district schools. These last conclusions can support the third paragraph´s idea.

```python
import pandas as pd
```

```python
#Reading the files 

schools = pd.read_csv('./Resources/schools_complete.csv')

students = pd.read_csv('./Resources/students_complete.csv')
```

```python
#Merging the tables 

df = students.merge(schools)
df
```

## District Summary

```python
#Total number of unique values


Nschools = len(df['school_name'].unique())
Nschools
```

```python
#Total students 
Nstudents = len(df['student_name'])
Nstudents
```

```python
#Total budgets
Tbud = schools['budget'].sum()
Tbud
```

```python
#Average math score 
avgmath = df['math_score'].mean()
avgmath
```

```python
#Average reading socre 
avgread = df['reading_score'].mean()
avgread
```

```python
#% of sutdents who passed math 
#Created a new df with the scores filtered higher or 
passmath = df[df['math_score']>=70]
permath = (len(passmath)/Nstudents)*100
permath 
```

```python
#% of students who passed reading 
#Created a new df with the scores filtered higher or = to 70
passread = df[df['reading_score']>=70]
perread = (len(passread)/Nstudents)*100
perread
```

```python
#% of students who passed math and reading 
passboth = df.loc[(df['math_score']>=70)& (df['reading_score']>=70)]
perboth = (len(passboth)/Nstudents)*100
perboth
```

```python
#Summary data frame 
district_summary = (pd.DataFrame([[Nschools,Nstudents,Tbud,avgmath,avgread,permath,perread,perboth]], 
             columns =['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score',
                      '% Passing Math','% Passing Reading', '% Overall Passing']))
#Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary
```

## School Summary

In every calculation I´m reseting the index and renameing the columns,
because I´m preparing the frames to be merge into the final, at the end.

```python
#School names 
schoolnames = pd.DataFrame(df['school_name'].unique())
```

```python
#school type 
schoolty = pd.DataFrame(df.groupby('school_name')['type'].unique()).reset_index().rename(columns={'school_name':'index','type':'School Type'})

schoolty['School Type'] = schoolty['School Type'].str[0]
```

```python
#Total students per school
N_perschool = pd.DataFrame(df['school_name'].value_counts()).reset_index().rename(columns={'school_name':'Total Students'})

#Creating the summary data frame 
per_school_summary = pd.merge(schoolty,N_perschool)
```

```python
# Total school budget
schoolbud = pd.DataFrame(df.groupby('school_name')['budget'].unique()).reset_index().rename(columns={'school_name':'index','budget':'Total School Budget'})
schoolbud['Total School Budget'] = schoolbud['Total School Budget'].astype(int)

#Per student dudget (The column is created directly in the summary data frame)
per_school_summary = pd.merge(per_school_summary,schoolbud)
per_school_summary['Per Student Budget']= per_school_summary['Total School Budget']/per_school_summary['Total Students']
```

```python
#Average math score 
avg_schoolmath = pd.DataFrame(df.groupby('school_name')['math_score'].mean()).reset_index().rename(columns={'school_name':'index','math_score':'Average Math Score'})
avg_schoolmath['Average Math Score'] = avg_schoolmath['Average Math Score'].astype(float)
```

```python
#Average reading score
avg_schoolread = pd.DataFrame(df.groupby('school_name')['reading_score'].mean()).reset_index().rename(columns={'school_name':'index','reading_score':'Average Reading Score'})
avg_schoolread['Average Reading Score'] = avg_schoolread['Average Reading Score'].astype(float)

#adding to the summary frame the average scores
per_school_summary = pd.merge(per_school_summary,avg_schoolmath)
per_school_summary = pd.merge(per_school_summary,avg_schoolread)
```

```python
#% of sutdents who passed math 
#Used the df (passmath) created above with the scores filtered higher or = to 70
passmath_school = pd.DataFrame(passmath.groupby('school_name')['math_score'].count()).reset_index().rename(columns={'school_name':'index'})

#adding the number column so I can do calculations
permath_school = pd.merge(passmath_school,N_perschool)

#adding the % column to the summary directly
per_school_summary['% Passing Math'] = (permath_school['math_score']/permath_school['Total Students'])*100
```

```python
#% of sutdents who passed reading
#Used the df (passread) created above with the scores filtered higher or = to 70
passread_school = pd.DataFrame(passread.groupby('school_name')['reading_score'].count()).reset_index().rename(columns={'school_name':'index'})


#adding the number column so I can do calculations
perread_school = pd.merge(passread_school,N_perschool)

#adding the % column to the summary directly
per_school_summary['% Passing Reading'] = (perread_school['reading_score']/perread_school['Total Students'])*100
```

```python
#% of students who passed math and reading 
#Used the df (passboth) created above with the scores filtered higher or = to 70
passboth_school = pd.DataFrame(passboth.groupby('school_name')['school_name'].count()).rename(columns={'school_name':'count'}).reset_index()
passboth_school = passboth_school.rename(columns={'school_name':'index'})
#adding the number column so I can do calculations
perrboth_school = pd.merge(passboth_school,N_perschool)

#adding the % column to the summary directly
per_school_summary['% Overall Passing'] = (perrboth_school['count']/perrboth_school['Total Students'])*100
```

```python
#Summary data frame
#the summary data frame is already created
#set the index 
per_school_summary = per_school_summary.set_index('index')

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)


# Display the DataFrame
per_school_summary.index.name= None
per_school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
#Sort the schools by % Overall Passing in descending 

top_schools = per_school_summary.sort_values('% Overall Passing', ascending = False)
top_schools.head(5)
```

## Lowest-Performing Schools (by % Overall Passing)

```python
#Sort the schools by % Overall Passing in ascending order

bottom_schools = per_school_summary.sort_values('% Overall Passing', ascending = True)
bottom_schools.head(5)
```

## Math Scores by Grade

```python
#DataFrame that lists the average math score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#dataframes of each grade
nine = df[df['grade']=="9th"]
ten = df[df['grade']=="10th"]
eleven = df[df['grade']=="11th"]
twelve = df[df['grade']=="12th"]


#average of each grade per school
nine_avg =  pd.DataFrame(nine.groupby('school_name')['math_score'].mean()).reset_index()
ten_avg =  pd.DataFrame(ten.groupby('school_name')['math_score'].mean()).reset_index()
eleven_avg =  pd.DataFrame(eleven.groupby('school_name')['math_score'].mean())
twelve_avg =  pd.DataFrame(twelve.groupby('school_name')['math_score'].mean())

#creating the data frame 

math_scores_by_grade = (nine_avg.merge(ten_avg, on='school_name')
                .rename(columns={'math_score_x':'9th','math_score_y':'10th'})
                .merge(eleven_avg, on='school_name')
                .merge(twelve_avg, on='school_name')
                .rename(columns={'math_score_x':'11th','math_score_y':'12th'})
                .set_index('school_name')
                )

math_scores_by_grade.index.name = None
math_scores_by_grade
```

## Reading Scores by Grade

```python
#DataFrame that lists the average reading score for students of each grade level

#using the above dataframes for separating each grade

#average of each grade per school
nine_avgr =  pd.DataFrame(nine.groupby('school_name')['reading_score'].mean()).reset_index()
ten_avgr =  pd.DataFrame(ten.groupby('school_name')['reading_score'].mean()).reset_index()
eleven_avgr =  pd.DataFrame(eleven.groupby('school_name')['reading_score'].mean())
twelve_avgr =  pd.DataFrame(twelve.groupby('school_name')['reading_score'].mean())

#creating the data frame 

reading_scores_by_grade = (nine_avgr.merge(ten_avgr, on='school_name')
                .rename(columns={'reading_score_x':'9th','reading_score_y':'10th'})
                .merge(eleven_avgr, on='school_name')
                .merge(twelve_avgr, on='school_name')
                .rename(columns={'reading_score_x':'11th','reading_score_y':'12th'})
                .set_index('school_name')
                )

reading_scores_by_grade.index.name = None
reading_scores_by_grade
```

## Scores by School Spending

```python
#values
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary
school_spending_df['Per Student Budget'] = (school_spending_df['Per Student Budget'].astype(str).str.replace(r'\D', '').astype(int))/100
```

```python
#categorize spending based on the bins.
school_spending_df['Spending Ranges (Per Student)'] = (pd.cut(school_spending_df['Per Student Budget']
                                                              , bins= spending_bins, labels=labels, include_lowest=True))
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].map("${:,.2f}".format)
school_spending_df
```

```python
#  Calculate averages for the desired columns

spending_math_scores = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()).reset_index()
spending_reading_scores = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()).reset_index()
spending_passing_math = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()).reset_index()
spending_passing_reading = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()).reset_index()
overall_passing_spending = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()).reset_index()
```

```python
#Creating the dataframe

spending_summary = (spending_math_scores.merge(spending_reading_scores, on='Spending Ranges (Per Student)')
                  .merge(spending_passing_math, on='Spending Ranges (Per Student)')
                  .merge(spending_passing_reading, on='Spending Ranges (Per Student)')
                  .merge(overall_passing_spending, on='Spending Ranges (Per Student)')
                  .set_index('Spending Ranges (Per Student)')
                   )
                
spending_summary
```

## Scores by School Size

```python
#bins
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#adding the column 'school size'
per_school_summary['School Size'] = pd.cut(per_school_summary['Total Students'], bins= size_bins, labels=labels, include_lowest=True)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = pd.DataFrame(per_school_summary.groupby(["School Size"])["Average Math Score"].mean()).reset_index()
size_reading_scores = pd.DataFrame(per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()).reset_index()
size_passing_math = pd.DataFrame(per_school_summary.groupby(["School Size"])["% Passing Math"].mean()).reset_index()
size_passing_reading = pd.DataFrame(per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()).reset_index()
size_overall_passing = pd.DataFrame(per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()).reset_index()
```

```python
#Creating the data frame
size_summary = (size_math_scores.merge(size_reading_scores, on='School Size')
                  .merge(size_passing_math, on='School Size')
                  .merge(size_passing_reading, on='School Size')
                  .merge(size_overall_passing, on='School Size')
                  .set_index('School Size')
                   )

size_summary
```

## Scores by School Type

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = pd.DataFrame(per_school_summary.groupby(["School Type"])["Average Math Score"].mean()).reset_index()
average_reading_score_by_type = pd.DataFrame(per_school_summary.groupby(["School Type"])["Average Reading Score"].mean()).reset_index()
average_percent_passing_math_by_type = pd.DataFrame(per_school_summary.groupby(["School Type"])["% Passing Math"].mean()).reset_index()
average_percent_passing_reading_by_type = pd.DataFrame(per_school_summary.groupby(["School Type"])["% Passing Reading"].mean()).reset_index()
average_percent_overall_passing_by_type = pd.DataFrame(per_school_summary.groupby(["School Type"])["% Overall Passing"].mean()).reset_index()
```

```python
#Creating the data frame
type_summary = (average_math_score_by_type.merge(average_reading_score_by_type, on='School Type')
                  .merge(average_percent_passing_math_by_type, on='School Type')
                  .merge(average_percent_passing_reading_by_type, on='School Type')
                  .merge(average_percent_overall_passing_by_type, on='School Type')
                  .set_index('School Type')
                   )

type_summary
```
